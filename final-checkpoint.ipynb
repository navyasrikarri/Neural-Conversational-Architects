```python
vectorizer = TfidfVectorizer()
classifier = LogisticRegression(max_iter=10000)
```

```python
x = vectorizer.fit_transform(patterns)
classifier.fit(x, labels)
```

```python
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def chatbot_response(query):

    vectorized_query = vectorizer.transform([query])

    if max(classifier.predict_proba(vectorized_query)[0]) > 0.6:  # Confidence threshold
        intent = classifier.predict(vectorized_query)[0]

        for i in intents:
            if i["tag"] == intent:
                return random.choice(i["responses"])
    else:

        inputs = tokenizer.encode(query + tokenizer.eos_token, return_tensors="pt")
        outputs = model.generate(inputs, max_length=100, pad_token_id=tokenizer.eos_token_id)
        return tokenizer.decode(outputs[:, inputs.shape[-1]:][0], skip_special_tokens=True)
```

```python
print("Chatbot is ready! Type 'exit' to quit.")
while True:
    query = input("User-> ")
    if query.lower() == "exit":
        print("Chatbot-> Goodbye! Take care!")
        break
    print(f"Chatbot-> {chatbot_response(query)}")
```
